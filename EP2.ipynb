```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_excel('/content/Employees.xlsx')
```

```python
df.dropna(inplace=True)
```

```python
df['Overtime Hours']=df['Overtime Hours'].astype('int')
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Monthly Salary'].sum().sort_values(by='Monthly Salary',ascending=False)
sns.barplot(x='Gender',y='Monthly Salary',data=sales_gen)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Annual Salary'].sum().sort_values(by='Annual Salary',ascending=False)
sns.barplot(x='Gender',y='Annual Salary',data=sales_gen)
```

```python
sales_gen=df.groupby(['No'],as_index=False)['Monthly Salary'].sum().sort_values(by='Monthly Salary',ascending=False)
sns.barplot(x='No',y='Monthly Salary',data=sales_gen)
```

```python
sales_gen=df.groupby(['No'],as_index=False)['Annual Salary'].sum().sort_values(by='Annual Salary',ascending=False)
sns.barplot(x='No',y='Annual Salary',data=sales_gen)
```

```python
sales_age=df.groupby(['Years'],as_index=False)['Monthly Salary'].sum().sort_values(by='Monthly Salary',ascending=False)
sns.barplot(x='Years',y='Monthly Salary',data=sales_age)
```

```python
sales_age=df.groupby(['Years'],as_index=False)['Annual Salary'].sum().sort_values(by='Annual Salary',ascending=False)
sns.barplot(x='Years',y='Annual Salary',data=sales_age)
```

```python
sales_state=df.groupby(['Country'],as_index=False)['Monthly Salary'].sum().sort_values(by='Monthly Salary',ascending=False).head(5)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Country',y='Monthly Salary')
```

```python
sales_state=df.groupby(['Country'],as_index=False)['Annual Salary'].sum().sort_values(by='Annual Salary',ascending=False).head(5)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Country',y='Annual Salary')
```

```python
sales_state=df.groupby(['Department'],as_index=False)['Annual Salary'].sum().sort_values(by='Annual Salary',ascending=False).head(5)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Department',y='Annual Salary')
```

```python
sales_state=df.groupby(['Department'],as_index=False)['Annual Salary'].sum().sort_values(by='Annual Salary',ascending=False).head(5)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Department',y='Annual Salary')
```
